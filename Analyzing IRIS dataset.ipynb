## The Iris dataset 
- This was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.
- It includes three iris species with 50 samples each as well as some properties about each flower
- One flower species is linearly separable from the other two, but the other two are not linearly separable from each other
- Species of a flower is classified on the basis of sepal_length, sepal_width, petal_length and petal_width

```python
# importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# loading the dataset
iris=sns.load_dataset('iris')
```

## Univariate Analysis

### Histograms

```python
sns.distplot(iris['sepal_length'])
plt.show()
```

```python
sns.distplot(iris['sepal_width'])
plt.show()
```

```python
sns.distplot(iris['petal_length'])
plt.show()
```

```python
sns.distplot(iris['petal_width'])
plt.show()
```

### Inference from histograms
- sepal_length and sepal_width form a normal curve
- sepal_length has a median value of around 6, and sepal_width of 3
- petal_length and petal_width do not follow a normal curve

### Area plot

```python
iris.plot.area()
plt.show()
```

### Inference from areaplot
- petal_width is the smallest among the four
- sepal_length is larger than others
- petal_length has two different range of values

### Densityplot

```python
sns.distplot(iris['sepal_length'])
sns.distplot(iris['sepal_width'])
sns.distplot(iris['petal_length'])
sns.distplot(iris['petal_width'])
plt.show()
```

### Inference from densityplot
- petal_width is the smallest
- sepal_width is bigger than petal_width, but smaller than petal_length and sepal_length; sepal_width is spread across a narrow range
- petal_length has a wider range, compared to sepal_length, but is smaller


### boxplots to identify outliers

```python
sns.boxplot(data=iris)
plt.show()
```

### Inference from boxplots
- sepal_length has the largest dimensions; it has no outliers
- petal_length is smaller than sepal_length, but has a wider range of values; no outliers
- sepal_width is bigger than petal_width; it has a few outliers
- petal_width is smaller than all other dimensions; no outliers

### Horizontal Boxplots

```python
sns.boxplot(data=iris, orient='h')
plt.show()
```

### Inference from horizontal boxplots
- sepal_length is bigger than other dimensions
- sepal_width has a narrow range; but has some outliers
- petal_length has a much wider range of values
- petal_width has the smallest dimensions

### Violin plots

```python
sns.violinplot(data=iris)
plt.show()
```

### Inference from violin plots
- sepal_length and sepal_width values form a normal curve
- petal_length and petal_width values are concentrated in two separate nodes

### Horizontal Violin plots

```python
sns.violinplot(data=iris, orient='h')
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='deep')
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris, palette='deep')
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='sepal_length', hue='species', data=iris, palette='deep')
plt.show()
```

```python
sns.scatterplot(x='petal_width', y='sepal_width', hue='species', data=iris, palette='deep')
plt.show()
```

### Inference from Scatterplots
- setosa can be easily differentiated from its various scatterplots; it has smaller petal_length, petal_width, and sepal_length; only its sepal_width is bigger than the other two species
- virginica and versicolor can be differentiated on the basis of their petal_length and petal_width, but not on the basis of their sepal_length and sepal_width
- there seems to be a correlation between petal_length and petal_width for all the species, as well as between petal_length and sepal_length
- but there is no significant correlation between sepal_lenth and sepal_width for virginica and versicolor species; setosa shows a correlation

## Bivariate Analysis

### Pair plots

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

### Inference from Pairplots
- the three species cannot be differentiated from their sepal_length or sepal_width
- however, petal_length and petal_width is more useful in differentiating the species, particularly for setosa
- petal_length and petal_width have a distinct correlation for all the species
- sepal_length and sepal_width do not have as distinct a correlation

### Joint plot

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter')
plt.show()
```

## Inference from jointplot
- petal_length and petal_width have a distinct correlation
- sepal_length and sepal_width do not have as distinct a correlation

### Heatmap

```python
sns.heatmap(iris.corr(), annot=True, cmap='Reds')
plt.show()
```

### Inference from heatmap
- petal_length and petal_width have a very high correlation
- sepal_length and sepal_width have a negligible correlation
- sepal_length also have a high correlation with petal_length and petal_width
- but sepal_width has very low correlation with other dimensions

## Conclusion
- setosa species can be separated from the other two species with data on petal_length and petal_width
- versicolor and virginica is lesser distinguishable; however, petal_length and petal_width can still be used to differentiate between the two
