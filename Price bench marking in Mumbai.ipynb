## Price benchmarking in Mumbai for Food and HPC SKUs

### Analysing price competitiveness for price elastic SKUs
- with D Mart Ready
- with Jio Mart
- with Grofers
- with Big Basket

```python
# importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# uploading the price dataset
df=pd.read_excel('Mumbai Prices.xlsx')
```

```python
# reading the dataset
df.head()
```

```python
df.info()
```

### Comparing with D-Mart

```python
df_dmart=df.loc[(df['Big Bazaar']>1) & (df['D Mart ready']>1)]
sns.countplot(x='vs D Mart', data=df_dmart)
plt.show()
print("% of SKUs in which we are off-price compared to D Mart: ",(df_dmart['vs D Mart'].loc[df_dmart['vs D Mart']==True].count())/(len(df_dmart)))
```

```python
df_dmartprice = (df_dmart['vs D Mart'] == True).groupby(df_dmart['Class Description']).agg(lambda x: x.sum() / len(x)).reset_index(name='Off-Price Index')
sns.barplot(x='Class Description', y='Off-Price Index', data=df_dmartprice)
plt.xticks(rotation=90)
plt.show()
```
